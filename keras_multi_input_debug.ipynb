# Debugging Keras Multi-Input Error
Notebook ini menunjukkan bagaimana mendeteksi dan memperbaiki error ketika model Keras mengharapkan beberapa input tetapi hanya menerima satu.

Kesalahan serupa dengan yang muncul di aplikasi:
```
ValueError: Layer "functional_2" expects 2 input(s), but it received 1 input tensors.
```

## 1. Import Library yang Diperlukan

Kita akan mengimpor TensorFlow/Keras, NumPy, dan modul lain yang diperlukan.

````python
```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
```
````

## 2. Buat Model Multi-Input Sederhana

Model berikut memiliki dua input: satu berukuran (None, 100) dan satu lagi (None, 10).

````python
```python
# definisi model
input_a = layers.Input(shape=(100,), name='input_a')
input_b = layers.Input(shape=(10,), name='input_b')

x = layers.Dense(32, activation='relu')(input_a)
y = layers.Dense(32, activation='relu')(input_b)

concat = layers.concatenate([x, y])
output = layers.Dense(1, activation='sigmoid')(concat)

multi_model = models.Model(inputs=[input_a, input_b], outputs=output)
```
````

## 3. Mempersiapkan Data Masuk (Input)

Kita membuat hanya satu array untuk mensimulasikan penggunaan input tunggal.


````python
```python
# hanya menyediakan satu input
input_data = np.random.rand(1, 100).astype('float32')
```
````

## 4. Menjalankan Prediksi dengan Input Salah

Mencoba memanggil prediksi dengan hanya satu tensor dan menunjukkan error.

````python
```python
try:
    multi_model.predict(input_data)
except Exception as e:
    print("Terjadi error:", e)
```
````

````python
## 5. Mengecek Ringkasan (summary) dan Spesifikasi Input Model

Gunakan `model.summary()` dan `model.inputs` untuk melihat jumlah input dan bentuknya.

```python
print("Inputs:", multi_model.inputs)
print("Input shapes:", [inp.shape for inp in multi_model.inputs])
multi_model.summary()
```
````

## 5. Mengecek Ringkasan (summary) dan Spesifikasi Input Model

````python
```python
print("Inputs:", multi_model.inputs)
print("Input shapes:", [inp.shape for inp in multi_model.inputs])
multi_model.summary()
```
````

````python
## 6. Mengoreksi Bentuk Input dan Menjalankan Prediksi

Siapkan dua array sesuai asumsi dan jalankan prediksi.

```python
input_b = np.random.rand(1, 10).astype('float32')
result = multi_model.predict([input_data, input_b])
print("Prediksi berhasil, output shape", result.shape)
```
````

## 7. Tips Debugging Model dengan Banyak Input

- Pastikan urutan input ketika memanggil `predict` sama dengan urutan input saat model dibangun.
- Gunakan `model.input_shape` atau `model.inputs` untuk memeriksa bentuk yang diharapkan.
- Visualisasikan arsitektur dengan `tf.keras.utils.plot_model(model, show_shapes=True)`.
- Validasi data sebelum `predict` dan tangkap exception untuk memberikan pesan informatif.
